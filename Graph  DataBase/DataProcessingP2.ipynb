### Police Area values issue

```python
# manually collect zones from the pattern of beats
def get_zone(beat):
    if beat.startswith('1'):
        return 'Zone1'
    elif beat.startswith('2'):
        return 'Zone2'
    elif beat.startswith('3'):
        return 'Zone3'
    elif beat.startswith('4'):
        return 'Zone4'
    elif beat.startswith('5'):
        return 'Zone5'
    elif beat.startswith('6'):
        return 'Zone6'
    elif beat.startswith('7'):
        return 'Zone7'
    elif beat.startswith('0'):
        return 'Zone7'
    else:
        return 'Unknown'
```

```python
crime_1000['beat'] = crime_1000['beat'].astype(str)
crime_1000['Zone'] = crime_1000['beat'].apply(get_zone)
```

## Adjacent Zones Relationship

To answer some of the query:  Find the zones that have are adjacent and sharing the same high crime months. 

We need to find out the adjacent zones for each zone. 

```python
# Create AdjacentZones DataFrame
columns = ['ZoneKey', 'AdjacentZoneKey']
AdjacentZone = pd.DataFrame(columns=columns)

# Use adjacency_dict and zone_key_map
adjacency_dict = {
    'Zone1': ['Zone2', 'Zone3', 'Zone4', 'Zone5'],
    'Zone2': ['Zone1', 'Zone5', 'Zone6'],
    'Zone3': ['Zone1', 'Zone4', 'Zone5', 'Zone6'],
    'Zone4': ['Zone1', 'Zone3'],
    'Zone5': ['Zone1', 'Zone2', 'Zone3', 'Zone6'],
    'Zone6': ['Zone2', 'Zone3', 'Zone5']
}

zone_key_map = {
    'Zone1': 1,
    'Zone2': 2,
    'Zone3': 3,
    'Zone4': 4,
    'Zone5': 5,
    'Zone6': 6
}

# Add adjacency data into AdjacentZones DataFrame
for zone, adjacent_zones in adjacency_dict.items():
    zone_key = zone_key_map[zone]
    for adjacent_zone in adjacent_zones:
        adjacent_zone_key = zone_key_map[adjacent_zone]
        AdjacentZone = AdjacentZone.append({'ZoneKey': zone_key, 'AdjacentZoneKey': adjacent_zone_key}, ignore_index=True)

# Check AdjacentZones DataFrame
AdjacentZone.head()
```
